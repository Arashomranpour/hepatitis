```python
from sklearn.neighbors import LocalOutlierFactor
lof=LocalOutlierFactor(n_neighbors=5,contamination=0.05)
x=df
y=lof.fit_predict(x)
```

```python
np.unique(y,return_counts=True)
```

```python
lof_results = pd.DataFrame({'Outlier_Label': y})
df_with_labels = pd.concat([df, lof_results], axis=1)
outliers_df = df_with_labels[df_with_labels['Outlier_Label'] == -1]
outliers_df
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
IQR
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x="alk_phosphate")
```

```python
df_no_outlier=df[~ ((df<(q1-1.5*IQR))|(df>(q3+1.5*IQR))).any(axis=1)]
```

```python
print(df_no_outlier.shape)
print(df.shape)
```

```python
df_no_outlier.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
y=df["class"]
x=df.drop(columns=["class"])
```

```python
skb=SelectKBest(score_func=chi2,k=10)
best_feature_fit=skb.fit(x,y)
best_feature_fit.scores_
best_feature_transform=best_feature_fit.transform(x)
```

```python
feature_score=pd.DataFrame(best_feature_fit.scores_,columns=["feature_score"])
```

```python
x.columns
```

```python
a=['age', 'sex', 'steroid', 'antivirals', 'fatigue', 'malaise', 'anorexia',
       'liver_big', 'liver_firm', 'spleen_palpable', 'spiders', 'ascites',
       'varices', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime',
       'histology']
scores=pd.DataFrame({"names":a,"scores":feature_score.feature_score})
```

```python
scores.sort_values("scores",ascending=False)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
rfe=RFE(lr,n_features_to_select=8)
rfe_fit=rfe.fit(x,y)
```

```python
selected=pd.DataFrame(rfe_fit.support_,columns=["selected"])
ranking=pd.DataFrame(rfe_fit.ranking_,columns=["ranking"])
```

```python
rfe_feature=pd.concat([scores.names,selected,ranking],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

```

```python
xmodel=df[['protime', 'sgot', 'bilirubin', 'age', 'alk_phosphate', 'albumin',
       'spiders', 'histology', 'malaise', 'fatigue', 'ascites', 'varices']]
ymodel=y.copy()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xmodel,ymodel,test_size=0.3,random_state=7)
```

```python
xtrain_b,xtest_b,ytrain_b,ytest_b=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
lrg=LogisticRegression()
lrg.fit(xtrain,ytrain)
ypred=lrg.predict(xtest)
accuracy_score(y_pred=ypred,y_true=ytest)
```

```python
lrg=LogisticRegression()
lrg.fit(xtrain_b,ytrain_b)
ypred_b=lrg.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)
```

```python
dt=DecisionTreeClassifier()
dt.fit(xtrain_b,ytrain_b)
ypred_b=dt.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)

len(xtrain_b.columns)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain_b,ytrain_b)
ypred_b=knn.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)
```

```python
rf=RandomForestClassifier()
rf.fit(xtrain_b,ytrain_b)
ypred_b=rf.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)
```

```python
import joblib
model_file=open("random_forest_model.pkl","wb")
joblib.dump(rf,model_file)
model_file.close()
```

```python
model_file=open("knn_model.pkl","wb")
joblib.dump(knn,model_file)
model_file.close()
```

```python
model_file=open("lrg_model.pkl","wb")
joblib.dump(lrg,model_file)
model_file.close()
```

```python
model_file=open("dt_model.pkl","wb")
joblib.dump(dt,model_file)
model_file.close()
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(xtrain_b,ytrain_b)
ypred_b=svm.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
gbc.fit(xtrain_b,ytrain_b)
ypred_b=gbc.predict(xtest_b)
accuracy_score(y_pred=ypred_b,y_true=ytest_b)
```

```python
model_file=open("gb_model.pkl","wb")
joblib.dump(gbc,model_file)
model_file.close()
```
